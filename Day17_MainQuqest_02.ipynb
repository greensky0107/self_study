< 구매금액( amt : Amount) 의 Z-score 구하기>

```python
# cc_num 컬럼의 값마다 데이터 개수를 계산합니다.

cc_df['cc_num'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt의 평균, 표준편차를 계산하여 amt_info에 저장합니다.

amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean','std']).reset_index()
amt_info
```

```python
# amt_info data 를 pickle 형식 file로 저장
amt_info.to_pickle('./amt_info.pkl')
```

```python
#Question_05. cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장합니다.
# (left merge를 수행합니다.)

cc_df=cc_df.merge(amt_info,on='cc_num',how='left')
```

```python
#cc_df 확인
cc_df
```

```python
# Question_06. 결제금액(amt)의 z-score를 계산하여 amt_z 컬럼에 저장합니다.
# 평균: mean, 표준편차: std
# data 확인

cc_df['amt_z']=(cc_df['amt']-cc_df['mean'])/cc_df['std']
cc_df
```

```python
# 사기 거래 데이터(is_fraud=1)를 확인합니다. fraud이면 amt_z score가 높다.

cc_df[cc_df['is_fraud'] == 1]
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼을 제거합니다.
# data 확인
cc_df=cc_df.drop(['mean','std'], axis = 1)
cc_df
```

<Card Number,  Category 별 표준편차 구하기>

```python
# cc_num, category 컬럼의 값마다 결제금액(amt)의 평균, 표준편차를 계산하여 cat_info에 저장합니다.
cat_info=cc_df.groupby(['cc_num','category'])['amt'].agg(['mean','std']).reset_index()
cat_info
```

```python
#cat_info data frame을  pickle file 로 저장
cat_info.to_pickle('./cat_info.pkl')
```

```python
# Question_07. cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장합니다.
# (left merge를 수행합니다.)

cc_df=cc_df.merge(amt_info, on='cc_num', how='left')
```

```python
# Question_08. 결제금액(amt)의 z-score를 계산하여 cat_amt_z 컬럼에 저장합니다.
# 평균: mean, 표준편차: std
# cc_df 확인
cc_df['cat_amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']
cc_df
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼을 제거합니다.

cc_df=cc_df.drop(['mean','std'], axis =1)
cc_df
```

< 결제 시간 구해서,  아침,점심,저녁으로 그룹화하기>

```python
# Question_09. trans_date_trans_time 컬럼에서 시간 값을 추출하여 hour 컬럼에 저장합니다.
# (힌트: datetime 자료형을 이용합니다.)

cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
cc_df
```

```python
# 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수를 정의합니다.  

def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
# Question_10. hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장합니다.

cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
cc_df
```

```python
# Question_11. 시간대별로 데이터 수를 계산하여 출력합니다.

cc_df['hour_cat'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장합니다.

all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
all_cnt
```

```python
# cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장합니다.

hour_cnt = cc_df.groupby(['cc_num','hour_cat'])['amt'].count().reset_index()
hour_cnt
```

```python
# card holder가  시간대별 얼마나 결제했는 지 비중을 보려면,  hour_cnt 에다 all_cnt를 left join
# Question_12. cc_num, hour_cat 컬럼을 기준으로 cc_df와 hour_cnt 데이터를 합칩니다.
# (cc_df에서 left merge를 수행합니다.)

hour_cnt=hour_cnt.merge(all_cnt,on='cc_num',how='left')
```

```python
# Question_13. amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경합니다

hour_cnt=hour_cnt.rename({'amt_x': 'hour_cnt', 'amt_y': 'total_cnt'}, axis=1)
hour_cnt
```

```python
# Question_14. 전체 거래 건수(total_cnt)에 대한 시간대별 거래 건수(hour_cnt)의 비율을 계산하여 
# hour_perc 컬럼에 저장합니다.

hour_cnt['hour_perc']=hour_cnt['hour_cnt']/hour_cnt['total_cnt']
```

```python
#hour_cnt에서  칼럼 고르기
hour_cnt = hour_cnt[['cc_num','hour_cat','hour_perc']]
```

```python
# Question_15. cc_num, hour_cat 컬럼을 기준으로 cc_df와 hour_cnt 데이터를 합칩니다.
# (cc_df에서 left merge를 수행합니다.)

cc_df=cc_df.merge(hour_cnt,on=['cc_num','hour_cat'],how='left')
```

```python
#data 확인
cc_df.head()
```

```python
# 시간 관련 feature 분석이 완료되었으므로 불필요해진 컬럼들을 제거합니다.

cc_df=cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis =1)
```

```python
# Question_16. cc_num 컬럼의 값마다 distance의 평균, 표준편차를 계산하여 dist_info 변수에 저장합니다.

dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean', 'std']).reset_index()
```

```python
#distance info data frame을  pickle파일로 저장
dist_info.to_pickle('./dist_info.pkl')
```

```python
# Question_16. cc_num 컬럼을 기준으로 cc_df와 dist_info 데이터를 합칩니다.
# (cc_df에서 left merge를 수행합니다.)

cc_df=cc_df.merge(dist_info, on='cc_num', how='left')
```

```python
#data 확인
cc_df.head()
```

```python
# Question_17. distance 값의 z-score를 계산하여 dist_z 컬럼에 저장합니다.

cc_df['dist_z'] = (cc_df['distance'] - cc_df['mean']) / cc_df['std']
```

```python
# 불필요해진 컬럼들을 제거합니다.

cc_df=cc_df.drop(['lat','long','merch_lat','merch_long','mean','std'], axis = 1)
cc_df
```
