Step 2-1 : Data Processing  - 범주형 data

원산지와 살찔까요는 문자인데, 이를 숫자로 바꿔주기 (Lable Encoding, One Hot Encoding)

Step 2-1-a ;  Lable Encoding

```python
# 레이블(label) 인코딩 : fit 과 transform 명령어를 각각 사용.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['원산지'])
le.transform(df['원산지'])
```

```python
# fit_transform  :  fit과 transform을 한번에!
le.fit_transform(df['원산지'])
```

```python
# 데이터 변환
df['원산지'] = le.fit_transform(df['원산지'])
df
```

```python
# '살찔까요' column 도 레이블인코딩
le = LabelEncoder()
df['살찔까요'] = le.fit_transform(df['살찔까요'])
df
```

```python
#여러개의 columns를 한꺼번에 lable encoding

# 활용할 데이터 선택 - 원본 데이터 불러오기
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 1.데이터가 object 타입인 column 선택 ['원산지', '살찔까요']

# 1-1.여러개 columns를 동시에 방법1
cols = ['원산지', '살찔까요']

# 1-2.여러개 columns를 동시에 방법2
cols = df.select_dtypes(include='object').columns

# 2.for 반복문을 이용해서, 한 번에 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# 3.인코딩 결과 확인
df
```

Step 2-1-b ; One Hot Encoding  

각 카테고리별로 컬럼을 만들어서 해당되는 카테고리의 컬럼은 1, 나머지 컬럼은 0으로 인코딩

One Hot Encoding 한개의 column 변환

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
# 데이터프레임을 입력받기 때문에 대괄호를 2번 써야 합니다. (df[['원산지']])
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[['원산지']])
cat
```

```python
# 카테고리
ohe.categories_
```

```python
# 피처(컬럼)이름과 카테고리
ohe.get_feature_names_out()
```

```python
# 데이터프레임으로 변환
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
# 데이터 프레임 합치기
df = pd.concat([df,df_cat],axis=1)
df
```

```python
# 기존 컬럼 삭제
df = df.drop(['원산지'], axis=1)
df
```

One Hot Encoding 여러 개의 column 변환

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

Step 3 : Data Split  (데이터 분리)

1. train_test_split
-사이킷런의 train_test_split 라이브러리를 활용하면 쉽게 train set(학습데이터셋)과 test set (테스트셋)을 랜덤하게 나누어 준다.

2. test_size=0.3
-학습용으로 70%를 두고 30% 데이터는 테스트용으로 둔다.

3. random_state
-반복 연산시 일정하게 섞기 위해 사용하며 random값을 고정하여 동일한 결과를 얻을 수 있다.

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_df.drop('target', axis=1), diabetes_df['target'], test_size=0.3, random_state=1004)
```

```python
# train 데이터 확인(X)
X_train.head()
```

```python
# 학습 데이터 확인(y)
y_train.head()
```

```python
# 데이터 크기
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Step 4 : Decide Model (Linear Regression)

```python
# 머신러닝(회귀)
from sklearn.linear_model import LinearRegression

# 모델 선택
model = LinearRegression()
# 학습
model.fit(X_train, y_train)
# 예측
pred = model.predict(X_test)
pred
```

Step 5 : Evaluation (MSE mean_squared_error(실제값, 예측값))

```python
# 정확도 mean_squared_error(실제값, 예측값)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred)
```
